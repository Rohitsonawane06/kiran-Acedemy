```python
import pandas as pd
import numpy as np

student={
    "Name":["Rohit","Tejas","Vijay","Santosh","Harshal"],
    "Age":[21,np.nan,20,np.nan,19],
    "Marks":[89,90,78,87,99]
}
df=pd.DataFrame(student)
```

```python
# 3. DataFrame Basic Oprations
#1.Get the shape of the data frame
print(df.shape)
#2. get column name
print(df.columns)
#3.Get basic statistics
print(df.describe())
```

```python
#6 Filter student older than 20
df[df["Age"]>20]
```

```python
#7 Add new column 'Grade' based on marks
def add(g):
    if g>80:
        return 'A'
    elif g>65:
        return 'B'
    else:
        return 'C'
    
df["Grade"]=df["Marks"].apply(add)
df
```

```python
#8sort by age
df_sorted=df.sort_values(by="Age")
print(df_sorted)
```

```python
#9.Group by grade and get average age
df.groupby('Grade')['Age'].mean()
```

```python
#10.fill missing value with findout mean or forword filling or mode ffill()
df['Age'].fillna(df['Age'].mean(),inplace=True)

```

```python
#11.Show all data without NAN
cleandf=df.dropna()
print(cleandf)
```
